```python
"""Berechnung der Leistungsabgabe von PV Modulen."""

import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# ════════════════════════════════════════════════════════════
# PARAMETER – hier anpassen
# ════════════════════════════════════════════════════════════
lat         = 49.45   # Breitengrad  (Nürnberg)
lon         = 11.08   # Längengrad
neigung     = 30      # Modulneigung gegen Horizontale (Grad)
ausrichtung = 180     # Azimut der Modulnormalen (180° = Süd)
P_peak      = 1.0     # Anlagenleistung (kWp)
eta         = 0.80    # Systemwirkungsgrad

tage = [
    (datetime(2025,  3,  1),  1, "steelblue", "1. März"),
    (datetime(2025,  6,  1),  2, "goldenrod", "1. Juni"),
    (datetime(2025,  9,  1),  2, "tomato",    "1. September"),
    (datetime(2025, 12,  1),  1, "slategray", "1. Dezember"),
]

dt_h              = 5 / 60      # Zeitschritt in Stunden (5 Minuten)
umrechnungsfaktor = eta * P_peak  # kWh pro m²·h (bei G = G_STC)

# ════════════════════════════════════════════════════════════
# HILFSFUNKTIONEN
# ════════════════════════════════════════════════════════════

def julianisches_datum(dt):
    return (367 * dt.year
            - int(7 * (dt.year + int((dt.month + 9) / 12)) / 4)
            + int(275 * dt.month / 9)
            + dt.day + 1721013.5
            + (dt.hour + dt.minute / 60.0 + dt.second / 3600.0) / 24.0)

def sonnenposition_ra_dekl(JD):
    n   = JD - 2451545.0
    L   = (280.460 + 0.9856474 * n) % 360
    g   = np.radians((357.528 + 0.9856003 * n) % 360)
    lam = np.radians(L + 1.915 * np.sin(g) + 0.020 * np.sin(2 * g))
    eps = np.radians(23.439 - 0.0000004 * n)
    RA   = np.degrees(np.arctan2(np.cos(eps) * np.sin(lam), np.cos(lam))) / 15 % 24
    dekl = np.degrees(np.arcsin(np.sin(eps) * np.sin(lam)))
    return RA, dekl

def lmst(JD, lon_grad):
    T    = (JD - 2451545.0) / 36525.0
    GMST = (280.46061837
            + 360.98564736629 * (JD - 2451545.0)
            + 0.000387933 * T**2) % 360
    return (GMST + lon_grad) / 15 % 24

def horizontkoordinaten(datum, lat, lon, zeitzone):
    phi      = np.radians(lat)
    ortszeit = np.arange(0, 24.01, 5/60)
    hoehe_liste  = []
    azimut_liste = []
    for ot in ortszeit:
        utc    = ot - zeitzone
        dt_utc = datum + timedelta(hours=float(utc))
        JD       = julianisches_datum(dt_utc)
        RA, dekl = sonnenposition_ra_dekl(JD)
        delta    = np.radians(dekl)
        H        = np.radians((lmst(JD, lon) - RA) * 15)
        r = np.array([np.cos(delta) * np.cos(H),
                      np.cos(delta) * np.sin(H),
                      np.sin(delta)])
        R = np.array([[ np.sin(phi), 0, -np.cos(phi)],
                      [ 0,           1,  0           ],
                      [ np.cos(phi), 0,  np.sin(phi) ]])
        r_horiz = R @ r
        h = np.degrees(np.arcsin(np.clip(r_horiz[2], -1, 1)))
        A = (np.degrees(np.arctan2(r_horiz[1], r_horiz[0])) + 180) % 360
        hoehe_liste.append(h)
        azimut_liste.append(A)
    return ortszeit, np.array(hoehe_liste), np.array(azimut_liste)

def modul_normalenvektor(neigung_grad, ausrichtung_grad):
    neigung     = np.radians(neigung_grad)
    ausrichtung = np.radians(ausrichtung_grad)
    n = np.array([
         np.sin(neigung) * np.sin(ausrichtung),
         np.sin(neigung) * np.cos(ausrichtung),
         np.cos(neigung)
    ])
    return n / np.linalg.norm(n)

def einstrahlung(hoehe_grad, G0=1361, T=0.76):
    """
    Direkte Sonneneinstrahlung auf Erdoberfläche nach Luftmassenmodell.
    hoehe_grad: Sonnenhöhe in Grad
    T:          Transmissionskoeffizient (0.76 = klarer Tag)
    """
    if hoehe_grad <= 0:
        return 0.0
    sin_h = np.sin(np.radians(hoehe_grad))
    # Luftmasse nach Kasten & Young (robuste Näherung auch bei flachem Winkel)
    AM    = 1 / (sin_h + 0.50572 * (hoehe_grad + 6.07995) ** -1.6364)
    return G0 * T ** AM

def effektive_flaeche(hoehe_arr, azimut_arr, n_vec, A_modul=1.0):
    A_eff = []
    for h, A in zip(hoehe_arr, azimut_arr):
        if h <= 0:
            A_eff.append(0.0)
            continue
        h_rad = np.radians(h)
        A_rad = np.radians(A)
        s = np.array([
             np.cos(h_rad) * np.sin(A_rad),
             np.cos(h_rad) * np.cos(A_rad),
             np.sin(h_rad)
        ])
        cos_theta = max(0.0, np.dot(s, n_vec))
        # Einstrahlung auf das Modul = G(h) · cos(θ) · A_modul
        G = einstrahlung(h)
        A_eff.append(A_modul * cos_theta * G / 1000)  # normiert auf kW/m²
    return np.array(A_eff)

def stundliche_ertraege(ortszeit, A_eff):
    stunden    = np.arange(0, 24)
    ertrag_kwh = []
    for s in stunden:
        maske = (ortszeit >= s) & (ortszeit < s + 1)
        ertrag_kwh.append(np.sum(A_eff[maske]) * dt_h * umrechnungsfaktor)
    return stunden, np.array(ertrag_kwh)

# ════════════════════════════════════════════════════════════
# VORBERECHNUNGEN
# ════════════════════════════════════════════════════════════
n_vec = modul_normalenvektor(neigung, ausrichtung)
print(f"Normalenvektor Modul: {n_vec.round(3)}")

# Alle Tage vorausberechnen und speichern
ergebnisse = []
for datum, zeitzone, farbe, label in tage:
    ortszeit, hoehe, azimut = horizontkoordinaten(datum, lat, lon, zeitzone)
    A_eff                   = effektive_flaeche(hoehe, azimut, n_vec)
    stunden, ertrag_kwh     = stundliche_ertraege(ortszeit, A_eff)
    ergebnisse.append({
        "datum":      datum,
        "label":      label,
        "farbe":      farbe,
        "ortszeit":   ortszeit,
        "hoehe":      hoehe,
        "azimut":     azimut,
        "A_eff":      A_eff,
        "stunden":    stunden,
        "ertrag_kwh": ertrag_kwh,
    })

# ════════════════════════════════════════════════════════════
# PLOT 1+2: Höhenwinkel und Azimut
# ════════════════════════════════════════════════════════════
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13, 8), sharex=True)
fig.suptitle(f"Sonnenlauf – Nürnberg (φ={lat}°)", fontsize=13)

for e in ergebnisse:
    tag_maske = e["hoehe"] > 0
    ax1.plot(e["ortszeit"], e["hoehe"], color=e["farbe"], linewidth=2, label=e["label"])
    ax1.fill_between(e["ortszeit"], e["hoehe"], where=tag_maske, color=e["farbe"], alpha=0.08)
    ax2.plot(e["ortszeit"][tag_maske], e["azimut"][tag_maske], color=e["farbe"], linewidth=2, label=e["label"])

ax1.axhline(0, color="gray", linewidth=0.8, linestyle="--")
ax1.set_ylabel("Sonnenhöhe (Grad)")
ax1.set_ylim(-10, 90)
ax1.set_yticks(range(0, 91, 15))
ax1.grid(True, alpha=0.4)
ax1.legend(loc="upper left")
ax1.set_title("Höhenwinkel über dem Horizont")

ax2.axhline(180, color="gray",      linewidth=0.8, linestyle="--", label="Süd 180°")
ax2.axhline( 90, color="lightblue", linewidth=0.6, linestyle=":",  label="Ost 90°")
ax2.axhline(270, color="lightblue", linewidth=0.6, linestyle=":",  label="West 270°")
ax2.set_ylabel("Azimut (Grad, 0°=Nord)")
ax2.set_ylim(0, 360)
ax2.set_yticks([0, 90, 180, 270, 360])
ax2.set_yticklabels(["N 0°", "O 90°", "S 180°", "W 270°", "N 360°"])
ax2.grid(True, alpha=0.4)
ax2.legend(loc="upper left", ncol=2)
ax2.set_title("Azimut")
ax2.set_xlabel("Ortszeit (Stunden)")
ax2.set_xticks(range(0, 25, 2))
ax2.set_xticklabels([f"{h:02d}:00" for h in range(0, 25, 2)])

plt.tight_layout()
plt.show()

# ════════════════════════════════════════════════════════════
# PLOT 3: Effektive Querschnittsfläche
# ════════════════════════════════════════════════════════════
fig, ax = plt.subplots(figsize=(13, 4))
fig.suptitle(
    f"Effektiver Modulquerschnitt – Neigung {neigung}°, Süd | Nürnberg",
    fontsize=13)

for e in ergebnisse:
    ax.plot(e["ortszeit"], e["A_eff"], color=e["farbe"], linewidth=2, label=e["label"])
    ax.fill_between(e["ortszeit"], e["A_eff"], color=e["farbe"], alpha=0.08)

ax.set_ylabel("Effektive Fläche (m² pro m² Modul)")
ax.set_ylim(0, 1.05)
ax.set_yticks(np.arange(0, 1.1, 0.2))
ax.grid(True, alpha=0.4)
ax.legend(loc="upper left")
ax.set_xlabel("Ortszeit (Stunden)")
ax.set_xticks(range(0, 25, 2))
ax.set_xticklabels([f"{h:02d}:00" for h in range(0, 25, 2)])

plt.tight_layout()
plt.show()

# ════════════════════════════════════════════════════════════
# PLOT 4: Stündlicher PV-Ertrag (Säulendiagramm)
# ════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(14, 9), sharey=True)
fig.suptitle(
    f"Stündlicher PV-Ertrag – Nürnberg (φ={lat}°)\n"
    f"Anlage: {P_peak} kWp, η={eta}, Modul: {neigung}° Neigung, Süd",
    fontsize=13)

for idx, e in enumerate(ergebnisse):
    ax  = axes.flatten()[idx]
    tagessumme = e["ertrag_kwh"].sum()

    ax.bar(e["stunden"] + 0.5, e["ertrag_kwh"], width=0.8,
           color=e["farbe"], alpha=0.8, edgecolor="white", linewidth=0.5)
    ax.set_title(f"{e['label']}  (Tagesertrag: {tagessumme:.2f} kWh)")
    ax.set_xlim(0, 24)
    ax.set_xticks(range(0, 25, 2))
    ax.set_xticklabels([f"{h:02d}" for h in range(0, 25, 2)], fontsize=9)
    ax.set_ylabel("Ertrag (kWh)")
    ax.set_xlabel("Ortszeit (Stunden)")
    ax.grid(True, alpha=0.3, axis="y")

    for s, e_kwh in zip(e["stunden"], e["ertrag_kwh"]):
        if e_kwh > 0.005:
            ax.text(s + 0.5, e_kwh + 0.002, f"{e_kwh:.2f}",
                    ha="center", va="bottom", fontsize=7)

plt.tight_layout()
plt.show()

# ════════════════════════════════════════════════════════════
# VERGLEICHSTABELLE
# ════════════════════════════════════════════════════════════
print(f"\nPV-Anlage: {P_peak} kWp | Wirkungsgrad: {eta*100:.0f}% | Wolkenlos")
print(f"\n{'Datum':<16} {'Tagesertrag':>12} {'Spitzenleistung':>17}")
print("─" * 48)
for e in ergebnisse:
    tagessumme = e["ertrag_kwh"].sum()
    spitze     = e["A_eff"].max() * umrechnungsfaktor
    print(f"{e['label']:<16} {tagessumme:>10.2f} kWh  {spitze:>12.2f} kW")
```

```python
# Vergleich März vs. September im Detail
for idx, label in [(0, "1. März"), (2, "1. September")]:
    e = ergebnisse[idx]
    tag_maske = e["hoehe"] > 0
    ot_tag    = e["ortszeit"][tag_maske]
    idx_max   = np.argmax(e["hoehe"])

    print(f"── {label} ───────────────────────────────")
    print(f"  Sonnenaufgang   : {ot_tag[0]:.2f} Uhr (Ortszeit)")
    print(f"  Sonnenuntergang : {ot_tag[-1]:.2f} Uhr (Ortszeit)")
    print(f"  Tageslänge      : {ot_tag[-1]-ot_tag[0]:.2f} h")
    print(f"  Höchststand     : {e['hoehe'][idx_max]:.2f}° um {e['ortszeit'][idx_max]:.2f} Uhr")
    print(f"  Max. A_eff      : {e['A_eff'][idx_max]:.4f}")
    print(f"  Tagesertrag     : {e['ertrag_kwh'].sum():.3f} kWh")
    print()
```

```python
e_dez = ergebnisse[3]   # 1. Dezember

# ── Schritt 1: Sonnenhöchststand ──────────────────────────
idx_max = np.argmax(e_dez["hoehe"])
print(f"Sonnenhöchststand am 1. Dezember:")
print(f"  Uhrzeit : {e_dez['ortszeit'][idx_max]:.2f} Uhr")
print(f"  Höhe    : {e_dez['hoehe'][idx_max]:.2f}°")
print(f"  Azimut  : {e_dez['azimut'][idx_max]:.2f}°")
print(f"  A_eff   : {e_dez['A_eff'][idx_max]:.4f} m²")
print()

# ── Schritt 2: Tageslänge ─────────────────────────────────
tag_maske = e_dez["hoehe"] > 0
ot_tag    = e_dez["ortszeit"][tag_maske]
print(f"Sonnenaufgang : {ot_tag[0]:.2f} Uhr")
print(f"Sonnenuntergang: {ot_tag[-1]:.2f} Uhr")
print(f"Tageslänge    : {ot_tag[-1] - ot_tag[0]:.2f} h")
print()

# ── Schritt 3: Integral händisch nachrechnen ──────────────
print(f"Tagesintegral A_eff (Summe aller 5-min-Werte × dt):")
summe = np.sum(e_dez["A_eff"]) * dt_h
print(f"  Σ A_eff · dt = {summe:.4f} m²·h")
print(f"  × Umrechnungsfaktor {umrechnungsfaktor:.2f} = {summe * umrechnungsfaktor:.4f} kWh")
print()

# ── Schritt 4: Plausibilitätscheck gegen Referenzwert ─────
# Faustregel: Im Dezember in Nürnberg ca. 1-2 kWh/kWp/Tag
print(f"Zum Vergleich:")
print(f"  Typischer Dezember-Ertrag Nürnberg laut Literatur: 0.5–1.5 kWh/kWp/Tag")
print()

# ── Schritt 5: Stündliche Aufschlüsselung ─────────────────
print(f"{'Stunde':>10} {'Max.Höhe':>10} {'Max.A_eff':>10} {'Ertrag kWh':>12}")
print("─" * 48)
gesamt = 0
for s in range(24):
    maske  = (e_dez["ortszeit"] >= s) & (e_dez["ortszeit"] < s + 1)
    ertrag = e_dez["ertrag_kwh"][s]
    if ertrag > 0 or e_dez["hoehe"][maske].max() > -5:
        max_h = e_dez["hoehe"][maske].max()
        max_a = e_dez["A_eff"][maske].max()
        gesamt += ertrag
        print(f"{s:02d}–{s+1:02d} Uhr  {max_h:>10.2f}°  {max_a:>10.4f}  {ertrag:>12.4f}")
print("─" * 48)
print(f"{'Gesamt':>10} {gesamt:>44.4f} kWh")
```

```python
e_dez = ergebnisse[3]   # 1. Dezember

print("5-Minuten-Werte rund um Sonnenaufgang (7–9 Uhr):")
print(f"{'Uhrzeit':>10} {'Höhe':>8} {'A_eff':>8} {'Ertrag/5min':>12}")
print("─" * 44)
for i, ot in enumerate(e_dez["ortszeit"]):
    if 7.0 <= ot <= 9.0:
        h     = e_dez["hoehe"][i]
        a     = e_dez["A_eff"][i]
        e_5min = a * dt_h * umrechnungsfaktor
        print(f"{ot:>10.2f}  {h:>8.2f}°  {a:>8.4f}  {e_5min:>12.5f} kWh")
```
